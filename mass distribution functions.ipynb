```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

def bernoulli_pmf(p,k):
    return p**k*(1-p)**(1-k)

p=0.6
k_values=np.array([0,1])
pmf = bernoulli_pmf(p,k_values)
print(pmf)
```

```python
plt.bar(k_values,pmf)
plt.xlabel('k')
plt.ylabel('p(x=k)')
plt.title('bernoulli distribution (p=0.6)')
plt.show
```

```python
# binomial distribution
def binomial(n,p,x):
    return (factorial(n)/(factorial(x)*factorial(n-x))) * (p ** x) * ((1-p)**(n-x))

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

n=10
p=0.5
x=range(n+1)
y=[binomial(n,p,i) for i in x]
print(y)
```

```python
bio = pd.DataFrame({'X':x,'P(X)':y})
bio.transpose()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Binomial Distribution')
plt.show()
```

```python
## geometric distribution
def geometric(p,x):
    return p*(1-p)**x-1

p = 0.5
x = range(1, 11)
y = [geometric(p, i) for i in x]
print(y)
```

```python
geometric=pd.DataFrame({'x':x,'p(x)':y})
geometric.transpose()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Geometric distribution')
plt.show()
```

```python
## hypergeometric distribution
def hypergeometric(N,n,K,k):
    return (comb(K,k)*comb(N-K,n-k))/comb(N,n)

def comb(n,k):
    return factorial(n)/(factorial(k)*factorial(n-k))

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

N = 100
K = 20
n = 10
x = range(n+1)
y = [hypergeometric(N, K, n, i) for i in x]
print(y)
    
```

```python
hypo = pd.DataFrame({'X':x,'P(X)':y})
hypo.transpose()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('hypergeometric distribution')
plt.show()
```

```python
# poisson distribution
def poisson(lmbda, k):
    return (lmbda ** k) * (2.71828 ** (-lmbda)) / factorial(k)

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

lmbda = 4
x = range(20)
y = [poisson(lmbda, i) for i in x]
print(y)
```

```python
poi = pd.DataFrame({'X':x,'P(X)':y})
poi.transpose()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Poisson distribution')
plt.show()
```

```python
# exponential distribution
def exponential(lmbda, x):
    result = 1
    for i in range(int(x)):
        result *= lmbda
    result *= 2.71828 ** (-lmbda * x)
    return result

lmbda = 0.5
x = [i / 10 for i in range(100)]
y = [exponential(lmbda, i) for i in x]
print(y)
```

```python
exp = pd.DataFrame({'X':x,'P(X)':y})
exp.transpose()
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Exponential distribution')
plt.show()
```
